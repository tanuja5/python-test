```python
#import required packages
from nsepy import get_history
import pandas as pd
from datetime import date
```

```python
#downloading data
Infy = get_history(symbol="INFY", start=date(2015,1,1),end=date(2016,1,31))

Tcs = get_history(symbol="TCS", start=date(2015,1,1),end=date(2016,1,31))

niftyIt = get_history(symbol="NIFTYIT", start=date(2015,1,1),end=date(2016,1,31), index=True)
```

```python
#calculating moving avg
mov_avg = pd.DataFrame()
import matplotlib.pyplot as plt

def mov_avgg(x,y,z):
    mov_avg[x] = pd.rolling_mean(y['Close'], z*7)
    mov_avg[x+"_close"+str(z)] = y['Close']
    ax1 = plt.subplot2grid((1,1), (0,0))
    mov_avg[[x, x+"_close"+str(z)]].plot(ax=ax1)
    plt.show()
    
```

```python
#4 week moving average
mov_avgg('Infy',Infy,4)
mov_avgg('Tcs',Tcs,4)
mov_avgg('niftyIt',niftyIt,4)
```

```python
#16 week moving average
mov_avgg('Infy',Infy,16)
mov_avgg('Tcs',Tcs,16)
mov_avgg('niftyIt',niftyIt,16)
```

```python
#52 week moving average
mov_avgg('Infy',Infy,52)
mov_avgg('Tcs',Tcs,52)
mov_avgg('niftyIt',niftyIt,52)
```

```python
import numpy as np
#1 if 10% differed, 0 if not
niftyIt['vol_shock'] = np.where((((niftyIt['Volume'].shift(1) - niftyIt['Volume'])/ niftyIt['Volume'])*100 > 10) | (((niftyIt['Volume'] - niftyIt['Volume'].shift(1))/ niftyIt['Volume'])*100 > 10), 1, 0)
#1 if greater, 0 if lesser
niftyIt['dir_vol_shock'] = np.where((((niftyIt['Volume'].shift(1) - niftyIt['Volume'])/ niftyIt['Volume'])*100 > 10), 1, 0)


```

```python
#1 if 10% differed, 0 if not
Infy['vol_shock'] = np.where((((Infy['Volume'].shift(1) - Infy['Volume'])/ Infy['Volume'])*100 > 10) | (((Infy['Volume'] - Infy['Volume'].shift(1))/ Infy['Volume'])*100 > 10), 1, 0)
#1 if greater, 0 if lesser
Infy['dir_vol_shock'] = np.where((((Infy['Volume'].shift(1) - Infy['Volume'])/ Infy['Volume'])*100 > 10), 1, 0)


```

```python
#1 if 10% differed, 0 if not
Tcs['vol_shock'] = np.where((((Tcs['Volume'].shift(1) - Tcs['Volume'])/ Tcs['Volume'])*100 > 10) | (((Tcs['Volume'] - Tcs['Volume'].shift(1))/ Tcs['Volume'])*100 > 10), 1, 0)
#1 if greater, 0 if lesser
Tcs['dir_vol_shock'] = np.where((((Tcs['Volume'].shift(1) - Tcs['Volume'])/ Tcs['Volume'])*100 > 10), 1, 0)

```

```python
#1 if 2% differed, 0 if not
niftyIt['price_shock'] = np.where((((niftyIt['Close'].shift(1) - niftyIt['Close'])/ niftyIt['Close'])*100 > 2) | (((niftyIt['Close'] - niftyIt['Close'].shift(1))/ niftyIt['Close'])*100 > 2), 1, 0)
#1 if greater, 0 is lesser
niftyIt['dir_price_shock'] = np.where((((niftyIt['Close'].shift(1) - niftyIt['Close'])/ niftyIt['Close'])*100 > 2), 1, 0)

#1 if 2% differed, 0 if not
Infy['price_shock'] = np.where((((Infy['Close'].shift(1) - Infy['Close'])/ Infy['Close'])*100 > 2) | (((Infy['Close'] - Infy['Close'].shift(1))/ Infy['Close'])*100 > 2), 1, 0)
#1 if greater, 0 is lesser
Infy['dir_price_shock'] = np.where((((Infy['Close'].shift(1) - Infy['Close'])/ Infy['Close'])*100 > 2), 1, 0)

#1 if 2% differed, 0 if not
Tcs['price_shock'] = np.where((((Tcs['Close'].shift(1) - Tcs['Close'])/ Tcs['Close'])*100 > 2) | (((Tcs['Close'] - Tcs['Close'].shift(1))/ Tcs['Close'])*100 > 2), 1, 0)
#1 if greater, 0 is lesser
Tcs['dir_price_shock'] = np.where((((Tcs['Close'].shift(1) - Tcs['Close'])/ Tcs['Close'])*100 > 2), 1, 0)


#1 if 2% differed, 0 if not
niftyIt['black_swan'] = np.where((((niftyIt['Close'].shift(1) - niftyIt['Close'])/ niftyIt['Close'])*100 > 2) | (((niftyIt['Close'] - niftyIt['Close'].shift(1))/ niftyIt['Close'])*100 > 2), 1, 0)
#1 if greater, 0 is lesser
niftyIt['dir_black_swan'] = np.where((((niftyIt['Close'].shift(1) - niftyIt['Close'])/ niftyIt['Close'])*100 > 2), 1, 0)

#1 if 2% differed, 0 if not
Infy['black_swan'] = np.where((((Infy['Close'].shift(1) - Infy['Close'])/ Infy['Close'])*100 > 2) | (((Infy['Close'] - Infy['Close'].shift(1))/ Infy['Close'])*100 > 2), 1, 0)
#1 if greater, 0 is lesser
Infy['dir_price_shock'] = np.where((((Infy['Close'].shift(1) - Infy['Close'])/ Infy['Close'])*100 > 2), 1, 0)

#1 if 2% differed, 0 if not
Tcs['price_shock'] = np.where((((Tcs['Close'].shift(1) - Tcs['Close'])/ Tcs['Close'])*100 > 2) | (((Tcs['Close'] - Tcs['Close'].shift(1))/ Tcs['Close'])*100 > 2), 1, 0)
#1 if greater, 0 is lesser
Tcs['dir_black_swan'] = np.where((((Tcs['Close'].shift(1) - Tcs['Close'])/ Tcs['Close'])*100 > 2), 1, 0)


```

```python
#1 if price shock without vol shock holds, else 0

niftyIt['price_no_volume'] = np.where(((niftyIt['vol_shock'] == 0) & (niftyIt['price_shock'] == 1)), 1, 0)
Infy['price_no_volume'] = np.where(((Infy['vol_shock'] == 0) & (Infy['price_shock'] == 1)), 1, 0)
Tcs['price_no_volume'] = np.where(((Tcs['vol_shock'] == 0) & (Tcs['price_shock'] == 1)), 1, 0)
```

```python
#Color timeseries between two volume shocks in a different color (Red)
p = figure(title="Volume shocks", y_axis_type="linear",x_axis_type='datetime')

p.scatter(niftyIt.index, niftyIt.vol_shock, legend="Nifty", line_color="red", line_width = 3)

p.scatter(Infy.index, Infy.vol_shock, legend="Infy", line_color = 'green', line_width = 3)

p.legend.location = "top_left"

p.xaxis.axis_label = 'Year'
p.yaxis.axis_label = 'Count'

output_file("multiline_plot.html", title="Multi Line Plot")

show(p)  # open a browser
```
